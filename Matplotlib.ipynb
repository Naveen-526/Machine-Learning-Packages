```python
time        =  np.arange(0, 10, 0.1);

amplitude   =  np.sin(time)

plt.plot(time, amplitude)

plt.title('Sine wave')

plt.xlabel('Time')

plt.ylabel('Amplitude = sin(time)')

plt.axhline(y=0, color='k')

plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)
plt.plot(x,np.sin(x))
plt.axhline(y=0, color = 'k')
plt.axvline(x=4*np.pi/2, color = 'k')
plt.grid()
```
